```python
df = pd.read_csv("keywords_raw.csv")
df.head()
```

```python
df_replace = df.fillna(0)
df_replace.head()
```

```python
# df_replace.loc[row(s), column(s)]
# df_replace.loc[:, [2016, 2012, 2008]]
# df_replace[[2016, 2012, 2008]]
df_replace['2008-16 Average']=df_replace[['2016', '2012', '2008']].mean(axis=1)
df_replace.head()
```

```python
df_top_new=df_replace.loc[(df_replace['2008']>0) & (df_replace['2012']>0) & (df_replace['2016']>0)]
# df_top_new.sort_values('2020', ascending=False).head(10)
df_top_new['Keyword Ocurrence Change']=df_top_new["2020"]/df_top_new["2008-16 Average"]
df_top_new.sort_values('Keyword Ocurrence Change', ascending=False).head(50)
```

```python
df_to_new.to_csv('clean_zero_2008-2016.csv')
```
