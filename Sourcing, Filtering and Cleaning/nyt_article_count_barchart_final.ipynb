```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv('nyt_raw_2008.csv')
fir=len(df1)
fir
```

```python
df2=pd.read_csv('nyt_raw_2012.csv')
sec=len(df2)
sec
```

```python
df3=pd.read_csv('nyt_raw_2016.csv')
thi=len(df3)
thi
```

```python
df4=pd.read_csv('nyt_raw_2020.csv')
fou=len(df4)
fou
```

```python
tot=fir+sec+thi+fou
```

```python
chart=[fir, sec, thi, fou]
ticks=['2008', '2012', '2016', '2020']
plt.figure(figsize=(15,7))
barlist=plt.bar(ticks, chart, width=.5)
plt.xticks(size=14)
plt.yticks(size=14)
plt.grid(axis='y')

barlist[0].set_color('brown')
barlist[1].set_color('green')
barlist[2].set_color('purple')
barlist[3].set_color('blue')
plt.xlabel('Years(March-July)', size=20)
plt.ylabel('Article Count', size=20)
plt.title('Presidential Election Years NYT Articles Counts', size=22)

plt.show()
```
