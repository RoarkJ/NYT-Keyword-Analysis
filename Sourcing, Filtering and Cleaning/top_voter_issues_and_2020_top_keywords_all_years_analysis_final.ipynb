```python
df_2016=pd.read_csv('Data/top_keywords_2016_keywords_counts.csv')
```

```python
df_2020=pd.read_csv('Data/top_keywords_2020_keywords_counts.csv')
```

```python
df_all_years=pd.concat([df_2008, df_2012, df_2016, df_2020])
```

```python
df_all_years.to_csv('Data/all_years_top_keywords_keywords_count_by_day.csv')
```

```python
ticks=[row['publication_date']  for index, row in df_all_years.iterrows() if '-15' in(row['publication_date'])]
plt.figure(figsize=(22,10))

#plt.grid(axis='y')

plt.plot(df_all_years['publication_date'], df_all_years['keywords_count'].rolling(window=3).mean())
plt.title('Topics Count 9\nPresidential Election Year Top Keywords Topics and Top Voter Issues Daily Count', size=18)
plt.xlabel('Year and Month', size=18)
plt.ylabel('Topics Count', size=18)
plt.xticks(ticks, rotation=90, size=18)
plt.yticks(size=18)
plt.xlim(-10, len(df_all_years['publication_date'])+10)
# plt.tight_layout()
plt.savefig('Images/presidential_election_year_top_keywords_topics_and_daily_count')
#plt.show()
```
